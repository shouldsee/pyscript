```python
from multiprocessing import *
import time
import sys
# def do_work(a)
#     time.sleep(.5)
#     return(-a)
    
def worker(worker_queue,done_queue):
    while not worker_queue.empty():
        (i,item) = worker_queue.get()
        out = do_work(**item);
        done_queue.put((i,out));
        print(timer.time()-tic,'done',i)
    print('worker stopped')
    sys.exit()
#     return 0
        
#         worker_queue.task_done()
def multiP(do_work,source,num_threads=cpu_count(),job_time = 1,index = 0,sort = 0):
    global workers
    worker_queue = Queue()
    done_queue = Queue()
    while not worker_queue.empty():
        print(woker_queue.get())
    qsize=len(source);  
    for i,item in enumerate(source):
        worker_queue.put((i,item))
    
    workers=[];
    for i in range(num_threads):
        t = Process(target=worker,args=(worker_queue,done_queue))
#         t.daemon = True
        t.start()
        workers.append(t)


    worker_queue.close()
    worker_queue.join_thread()

    tic = timer.time()
#     p.join(qsize * job_time)
#     looptime = qsize * job_time;
    while True:
        if workers:
            for i,p in enumerate(workers):
                t = timer.time() - tic;
                if p.is_alive():
                    if t >  job_time:
                        p.terminate()
                        p.join()
                        workers.pop(i)
                    else:
                        time.sleep(job_time-t)
#                         print('join' ,job_time - t,p)
#                         p.join(job_time - t)
                else:
                    workers.pop(i)
        else:
            break
    queue=done_queue
#     queue.close()
    output=[];
    while not queue.empty():
        try:
            job = queue.get_nowait()
            if index:
                output.append(job);
            else:
                output.append(job[1]);
        except:
            break

    if sort:
        output = sorted(output, key=lambda x:x[0],reverse=0)
#     print(done_queue.get())
#     print(output)
#     print(sorted_output)
    print('{:d} of {:d} jobs done in {:2.2f}s'.format(len(output),qsize,timer.time()-tic))
    return(output)
```
